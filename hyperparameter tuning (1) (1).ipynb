```python
# dataframe columns
c_X = pd.DataFrame(iris_data.data, columns = iris_data.feature_names)
```

```python
# STANDARD Scale the dataset
from sklearn.preprocessing import StandardScaler
```

```python
sc_c = StandardScaler()
```

```python
sc_c_X = sc_c.fit_transform(c_X)
```

```python
sc_c_X
```

```python
# MANUAL BASIS....
# SPLITTING OF THE DATA..... 80,20 RATION.....

from sklearn.model_selection import train_test_split
```

```python
# performing the split...
# RESAMPLING + HYPERPARAMETER....
# x = sc_c_X, Y = c_Y
c_X_train, c_X_test, c_Y_train, c_Y_test = train_test_split(sc_c_X,c_Y, test_size=0.2,random_state=43)
```

```python
# .... Ridge_regression by using GridSearchCV
from sklearn.linear_model import Ridge
from sklearn import model_selection
```

```python
# model for classification.....svm for the ..... 
from sklearn import svm
```

```python
# ....
# class....
# no inbuild ..... 
# 23 days...
# knn.... from sc,,,,,
```

```python
svm_clf = svm.SVC(probability=True)
```

```python
# manual.....
svm_clf
```

```python
svm_clf.
```

```python
params = {"C":[0.1,1,10,100,1000],
         "gamma":[1,0.1,0.01,0.001,0.0001],
          "kernel":['linear','rbf'] # poly,sigmoid
         }
```

```python
# cv.... k = 3
gs = model_selection.RandomizedSearchCV(n_iter=20,estimator =svm_clf,param_distributions = params,
                                        scoring = 'neg_log_loss',cv = 3, n_jobs = 1)
```

```python
gs.fit(c_X_train,c_Y_train) # when we are going to use cv.....k = 3... classification....
```

```python
best_model = gs.best_estimator_
```
