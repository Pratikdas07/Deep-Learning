```python
#Import the neccasary Libraries
import pandas as pd
import tensorflow as tf
tf.__version__
```

```python
cat_col = ['DOJ.Extended','Offered.band','Joining.Bonus','Candidate.relocate.actual','Gender','Candidate.Source','LOB','Location']
Hr_dummy = pd.get_dummies(Hr,columns=cat_col,drop_first=True)
```

```python
Hr_dummy.head()
```

```python
Hr_dummy.shape
```

# Splitting the data

```python
Y = Hr_dummy[['Status']]
X = Hr_dummy.drop(columns=['Status','SLNO','Candidate.Ref'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
len(x_train),len(x_test),len(y_train),len(y_test)
```

# Building the ANN
## Step 1: Initializing the ANN

```python
ann = tf.keras.models.Sequential()
```

# Step 2: Adding the input layer and first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=4,activation='relu'))
```

# Step 3: Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=4,activation='relu'))
```

# Step 4: Adding the output layer

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

# Compiling the ANN

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

# Training the ANN

```python
ann.fit(x_train,y_train,batch_size=32,epochs=50)
```

```python
y_test
```

```python
y_test['Prediction'] = ann.predict(x_test)
```
