```python
import tensorflow as tf
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

# Generating Images from our training and test set

```python
train_datagen = ImageDataGenerator(rescale = 1/255)
```

```python
train_set= train_datagen.flow_from_directory('../input/cat-and-dog/training_set/training_set',target_size=(64,64),batch_size=32,class_mode='binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1/255)
test_set = test_datagen.flow_from_directory('../input/cat-and-dog/test_set/test_set',target_size=(64,64),batch_size=32,
                                              class_mode='binary')
```

# Initialising the CNNÂ¶

```python
cnn = tf.keras.models.Sequential()
```

```python
# Convolution
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
```

```python
#Polling
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#Adding a second layer
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
```

```python
#Flattening:
cnn.add(tf.keras.layers.Flatten())
```

```python
#Full connection
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
#Output Layer and compile
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
#Compiling the CNN
cnn.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
#Running the CNN
cnn.fit(x=train_set,validation_data=test_set,epochs=7)
```

```python
import numpy as np
from keras.preprocessing import image
val_image = image.load_img('../input/cat-and-dog/test_set/test_set/cats/cat.4007.jpg',target_size=(64,64))
val_image = image.img_to_array(val_image)
val_image = np.expand_dims(val_image,axis=0)
```
